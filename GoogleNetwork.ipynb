```python
hc=function(g, file){
    cw=cluster_walktrap(g, modularity = TRUE, membership = TRUE)
    
    num_circle=length(readLines(file))
    df=read.delim(file, sep="\t", header=FALSE)
    N=sum(!is.na(df))-num_circle #exclude three circle names    
    HC=0
    for (i in 1:num_circle){
        rl=readLines(file)[i]
        srl=strsplit(rl, '\t')
        dfi=data.frame(matrix(unlist(srl), nrow=length(srl), byrow=T))
        Ai=as.vector(unlist(dfi[1,]))
        a=length(Ai)-1
        HC= HC+ a/N*log(a/N)
    }
    HC=-HC
    print(HC)
    
    B=sizes(cw)
    HK=0
    for (b in B){
        HK= HK+ b/N*log(b/N)
    }
    HK=-HK
    print(HK)
    
    HCK=0
    for (j in 1:length(cw)){
        b=B[j]
        for (i in 1:num_circle){
            rl=readLines(file)[i]
            srl=strsplit(rl, '\t')
            dfi=data.frame(matrix(unlist(srl), nrow=length(srl), byrow=T))
            Ai=as.vector(unlist(dfi[1,]))
            dfj=data.frame(communities(cw)[j])
            tdfj=data.frame(t(dfj))
            Aj=as.vector(unlist(tdfj[1,]))
            AjAi=intersect(Aj, Ai)
            Aji=length(AjAi)
            #print(Aji)
            if (Aji!=0){
                HCK= HCK+ Aji/N*log(Aji/b)
            }
        }
    }
    HCK=-HCK
    print(HCK)

    HKC=0
    for (i in 1:num_circle){
        rl=readLines(file)[i]
        srl=strsplit(rl, '\t')
        dfi=data.frame(matrix(unlist(srl), nrow=length(srl), byrow=T))
        Ai=as.vector(unlist(dfi[1,]))
        a=length(Ai)-1
        for (j in 1:length(cw)){
            dfj=data.frame(communities(cw)[j])
            tdfj=data.frame(t(dfj))
            Aj=as.vector(unlist(tdfj[1,]))
            AjAi=intersect(Aj, Ai)
            Aji=length(AjAi)
            #print(Aji)
            if (Aji!=0){
                HKC= HKC+ Aji/N*log(Aji/a)
            }
        }
    }
    HKC=-HKC
    print(HKC)
    
    h=1-HCK/HC
    c=1-HKC/HK
    hc=list('h'=h, 'c'=c)
    return(hc)
}
```

```python
hc(g109, './109327480479767108490.circles')
```

If the total number of people with circle information (N) is roughly the same as the total number of people with community information, like when number of true values is roughly the same as the number of predicted values, H(C)>H(C|K)>0, H(K)>H(K|C)>0, so 0<h<1 and 0<c<1. This is the case for node ID = 109327480479767108490

```python
hc(g115, './115625564993990145546.circles')
```

At node ID = 115625564993990145546, there are 31 circles and 25 communities. There are 6467 people who have circle information (N=6467) but only 923 people has community information. Therefore b is lot smaller than N, which could make H(K) very small. If Aji is just a bit smaller than bi but ai is a lot smaller than N, H(K|C) could be bigger than H(K), which makes H(K|C)/H(K)>1 and thus cause c<0

```python
hc(g101, './101373961279443806744.circles')
```

When H(C|K) is larger than H(C), which means conditional entropy of circles is larger than normal entropy at circles, h<0. There are 29 communities but only 3 circles. Conditional entropy considers all the communities, which introduce more noise than normal circle entropy.

When there are a lot of people in the community that don't have circle information, N could be smaller than b, thus b/N*log(b/N)>0, and thus H(K)<0. which is what happened in node ID= 101373961279443806744. Since H(K)<0, and H(K|C)>0, H(K|C)/H(K)<0, and thus c>1
